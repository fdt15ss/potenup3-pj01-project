```python
import pandas as pd

df = pd.read_excel(
    "data/2021년 2월부터 2026년 1월까지 사주,운세,타로 검색량 상대 수치.xlsx",
    header=6,
)
df
```

```python
import matplotlib.pyplot as plt
import koreanize_matplotlib


# 문자열 → datetime (새 컬럼으로)
df["날짜_dt"] = pd.to_datetime(df["날짜"], errors="coerce")

plt.figure()
for col in df.columns[1:-1]:  # 검색어 컬럼들만
    plt.plot(df["날짜_dt"], df[col], label=col)

plt.xlabel("Date")
plt.ylabel("Relative Search Volume")
plt.title("Search Trend Over Time")
plt.legend()
plt.show()
```
